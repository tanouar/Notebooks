```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Exemple de DataFrame
data = {
    'numero_page': [1, 2, 3, 4, 5],
    'type': ['A', 'B', 'A', 'C', 'B'],
    'INFO_CONSEIL-PRO_PENSEZ-Y_FOCUS': ['Conseil1', 'Conseil2', 'Conseil3', 'Conseil4', 'Conseil5']
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
domtom = df 

data = domtom.groupby("type").head()
data
```

```python
# Initialiser OneHotEncoder
encoder = OneHotEncoder()
```

```python
# Appliquer OneHotEncoder à la colonne 'type'
encoded_array = encoder.fit_transform(df[['type']])
```

```python
encoded_array
```

```python
# Obtenir les noms des colonnes résultantes
encoded_columns = encoder.get_feature_names_out(['type'])
encoded_columns
```

```python
encode_df = pd.DataFrame(encoded_array, columns=encoded_columns)
encode_df.head()
```

```python
encode_df = pd.get_dummies(data[["type", "INFO_CONSEIL-PRO_PENSEZ-Y_FOCUS"]])
```

```python
df_final = pd.concat([df.drop(columns=['type']), encode_df], axis=1, join="inner")
df_final
```

```python
result = pd.concat([df, encode_df], axis=1, join='inner')
result
```
